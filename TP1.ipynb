```python
import io
import matplotlib
import matplotlib.pyplot as plt
import numpy
import pandas as pd
import seaborn

seaborn.set_context('talk')
```

```python
url = 'https://cs.famaf.unc.edu.ar/~mteruel/datasets/diplodatos/sysarmy_survey_2020_processed.csv'
df = pd.read_csv(url)
```

```python
salary_col='salary_monthly_NETO'
studies_col='profile_studies_level'
```

```python
df[salary_col].describe().apply(lambda s: '{0:.2f}'.format(s))
```

```python
sub_df = df[[salary_col,studies_col]]
sub_df
```

Hiótis

```python
sorted_studies_levels = ['Primario', 'Secundario', 'Terciario', 'Universitario',
                         'Posgrado', 'Doctorado', 'Posdoctorado']
plt.figure(figsize=(12, 6))
seaborn.boxplot(data=sub_df, x=salary_col, y=studies_col,
                color='orangered', order=sorted_studies_levels)
plt.ticklabel_format(style='plain', axis='x')
```

A simple vista pareciera que PostDoctorado tiene mayor nivel de sueldo. Hay muchos valores extremos en la variable Universitario.
La distribución de los distintos niveles educativos está concentrada aproximadamente en los mismos valores.

```python
seaborn.catplot(x=studies_col,y=salary_col,data=sub_df,order=sorted_studies_levels,height=10)
```

La muestra esta sesgada a niveles educativos universtiaruio y terceriario. Se observa poca cantidad en Postdoctorado lo que dificulta el análisis.

```python
sub_df_2 = sub_df[sub_df[salary_col] < 400000]
seaborn.catplot(x=studies_col,y=salary_col,data=sub_df_2,order=sorted_studies_levels,height=10)
```

Tratamos de eliminar los valores extremos para visulizar de mejor manera la distribución de los salarios.
Podemos concluir que el aumento comienza en el nivel universitario y en los demás no influye.
Confirmamos la hipótesis planteada hasta cierto punto.
